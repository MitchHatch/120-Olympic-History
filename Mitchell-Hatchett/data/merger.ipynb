```python
import pandas as pd
```

```python
athlete_df = pd.read_csv ("athlete_events.csv")
noc_regions_df = pd.read_csv("noc_regions.csv")
world_cities_df = pd.read_csv('worldcities.csv')
```

```python
athlete_df.head()
```

```python
athlete_df.columns
```

```python
athlete_clean_up = athlete_df.rename(columns={"Team":"Country"})
athlete_clean_up.head()
```

```python
athlete_clean_up.count()
```

```python
athlete_clean_up = athlete_clean_up.dropna(how='any')
```

```python
athlete_clean_up.count()
```

```python
athlete_clean_up.head()
```

```python
athlete_clean_up.reset_index()
```

```python
athlete_by_year = athlete_clean_up.sort_values(by=['Year','Country'])
athlete_by_year.head()
```

```python
world_cities_df.head()
```

```python
world_cities_df = world_cities_df.drop(columns=['iso2','iso3','admin_name','id','population','city_ascii','capital'])
```

```python
world_clean_up_df = world_cities_df.rename(columns={"country":"Country", 'city':'City'})
world_clean_up_df.head()
```

```python
merged_df = pd.merge(athlete_by_year,world_clean_up_df, on = ['Country','City'])
merged_df.head()
```

```python
merged_df1 = merged_df.sort_values(by=['Country'])
merged_df1.head()
```

```python
age_groups = merged_df1.groupby(by= "Age",sort = True , group_keys=True)
print(age_groups)
```
